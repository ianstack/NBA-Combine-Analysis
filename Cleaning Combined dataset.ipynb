```python
# Remove columns from Bench Press to ON_MOVE_COLLEGE as there are a lot of null values and this data was not collected
# in early years of the NBA combine.

# Dropped these columns as there others that represent these: (Winspan,height with shoes, Standing Reach & Lane Agility time)
```

```python
columns = ['WINGSPAN_FT_IN', 'HEIGHT_WO_SHOES', 'HEIGHT_WO_SHOES_FT_IN', 
        'HEIGHT_W_SHOES_FT_IN', 'STANDING_REACH_FT_IN', 'MODIFIED_LANE_AGILITY_TIME', 'Unnamed: 0', 'THREE_QUARTER_SPRINT']

draft_df.drop(draft_df.iloc[:, 24:49], inplace=True, axis=1)

draft_df.drop(draft_df.columns[[draft_df.columns.get_loc(col) for col in columns if col in draft_df]], 
              axis=1, inplace=True)
```

```python
draft_df.head()
```

```python
draft_df.isnull().sum()
```

```python
# Percentage of missing values in each column of the nba_draft_data.csv
draft_df.isnull().mean().mul(100).round(2).sort_values(ascending=False)


draft_df.isnull().mean()[draft_df.isnull().mean() !=0].mul(100)\
.round(2).sort_values(ascending=False) 
```

```python
# Looking at the rest of the null values I plan on replacing them with the mean per that columna and per position

draft_df.groupby("POSITION")["HAND_LENGTH"].mean()
```

I decided to replace the below listed columns with the average per position in that column as this would
give the most similar data for this null's.  This will allow me to not skew the dataset as replacing each
per postion will be the most similar to each player.

```python
draft_df["HAND_LENGTH"].fillna(draft_df.groupby("POSITION")["HAND_LENGTH"].transform("mean"), inplace = True)
draft_df["HAND_WIDTH"].fillna(draft_df.groupby("POSITION")["HAND_WIDTH"].transform("mean"), inplace = True)
draft_df["BODY_FAT_PCT"].fillna(draft_df.groupby("POSITION")["BODY_FAT_PCT"].transform("mean"), inplace = True)
draft_df["HEIGHT_W_SHOES"].fillna(draft_df.groupby("POSITION")["HEIGHT_W_SHOES"].transform("mean"), inplace = True)
draft_df["STANDING_VERTICAL_LEAP"].fillna(draft_df.groupby("POSITION")["STANDING_VERTICAL_LEAP"].transform("mean"), inplace = True)
draft_df["MAX_VERTICAL_LEAP"].fillna(draft_df.groupby("POSITION")["MAX_VERTICAL_LEAP"].transform("mean"), inplace = True)
draft_df["LANE_AGILITY_TIME"].fillna(draft_df.groupby("POSITION")["LANE_AGILITY_TIME"].transform("mean"), inplace = True)


```
