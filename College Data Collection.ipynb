```python
# In order for this code to run, I needed to create a column in the data set, labeled title Link, containing
# the text string to be entered in to the College stats url.  The string is formattted as
# "[First name]-[last name]-1" for a majority of players.
# This allows me to accurately pull from the website for each player
```

```python
# Lowercase for first name as url only uses lowercase letters

draft_df['FIRST_NAME'] = draft_df['FIRST_NAME'].apply(str.lower)
```

```python
import sys
import numpy as np
np.set_printoptions(threshold=np.inf)
```

```python
# Look at all Last names for any unique changes that may mess up the collection function

draft_df['LAST_NAME'].unique()
```

```python
# Lower all last names as the url uses lowercases letter only

draft_df['LAST_NAME'] = draft_df['LAST_NAME'].str.lower()
```

```python
# Check previous work

draft_df.head()
```

```python
# Create new list with each player formatted correctly so it can be inserted into the url

name_list = []

for first_name, last_name in zip(draft_df['FIRST_NAME'], draft_df['LAST_NAME']):
    name_list.append(f'{first_name}-{last_name}-1')
```

```python
# Check prevous code work

name_list
```

```python
# Add name_list into a new column for collection function to grab from

draft_df['Link'] = name_list
```

```python
# Replace periods and question marks with blank spaces as some names have apostrophe's in their name. (ie. Jr., Sr.)
# This allows me to gather some more players

draft_df['Link'] = draft_df['Link'].str.replace(".", "", regex=False).str.replace("?", "", regex=False)
```

```python
# import time as I needed to create a sleepr functin to pull all college data from sports.reference.com, as the
# the website had a rate-limit per pull requests.  This allows me to pull data every 3 seconds (time.sleep(3))

import time
```

```python
# This block of code is a for loop that I used to pull all College W/s (Win Share) data from sports.reference.com

# First I create a new column for data to be stored in 
# Next I created a for loop to grab each value from the 'Link' column
# Next I utilized read_html to read in the url that I pulled
# From I converted this data into a dataframe as it is easier to work with
# Next I created an 'except' to create a null value for any player that it did not work with.

draft_df['College WS'] = ''

for player, idx in zip(draft_df['Link'], draft_df.index):
    try:
        x = pd.read_html(f'https://www.sports-reference.com/cbb/players/{player}.html', attrs = {'id': 'players_advanced'})
        x = pd.DataFrame(x[0])
        ws_val = x['WS'].iloc[-2]
        draft_df.loc[idx, 'College WS'] = ws_val
    except:
        draft_df.loc[idx, 'College WS'] = 'N/A'
    time.sleep(3)
```

```python
# Sace data as csv

draft_df.to_csv('./data/College_WS.csv')
```
